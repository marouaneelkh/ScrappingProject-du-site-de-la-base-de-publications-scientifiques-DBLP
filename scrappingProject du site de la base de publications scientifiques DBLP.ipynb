```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
data_list = []
URL = "https://dblp.org/pid/07/1501.html"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
tab_elements = soup.find_all ("cite", class_="data tts-content")

for i in range (len(tab_elements)):
    title = tab_elements[i].find("span", class_="title").get_text()
    date = tab_elements[i].find("span" , itemprop="datePublished").get_text()
    authors = [author.get_text() for author in tab_elements[i].find_all("span", itemprop="author")]

    data_list.append((title, date, authors))
```

```python
df = pd.DataFrame(data_list,columns=['article','Authors', 'Date'])
df
```
